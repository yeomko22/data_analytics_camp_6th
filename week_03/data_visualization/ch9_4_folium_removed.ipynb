### 지도 위에 마커 찍기

```python
import pandas as pd
```

```python
df = pd.read_excel("./data/seoul_universities.xlsx", engine="openpyxl")
```

```python
df = df.rename(columns={
    "Unnamed: 0": "name",
    "위도": "latitude",
    "경도": "longitude"
})
```

```python
for i in range(len(df)):
    row = df.iloc[i]
    name, latitude, longitude = row
    folium.Marker([latitude, longitude], popup=name).add_to(seoul_map)
```

```python
seoul_map
```

마커의 크기나 모양도 조정할 수 있습니다.

```python
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)
for i in range(len(df)):
    row = df.iloc[i]
    name, latitude, longitude = row
    folium.CircleMarker(
        [latitude, longitude], 
        radius=10,
        popup=name,
        color="brown",
        fill=True,
        fill_color="coral",
        fill_opacity=0.7
    ).add_to(seoul_map)
```

```python
seoul_map
```

### 지도 영역에 단계 구분도 표시하기
지역별로 구분지어 색상을 다르게 칠하는 시각화를 folium을 통해 구현해보겠습니다. 이를 위해선 각 지역별 경계를 나타내는 데이터와 지역에 매핑하고자 하는 데이터가 필요합니다.

```python
population_df = pd.read_excel("./data/gg_population.xlsx", engine="openpyxl", index_col="구분")
```

```python
population_df.head()
```

```python
gg_map = folium.Map(location=[37.5502, 126.982], zoom_start=10)
```

```python
target_year = 2017
```

```python
import json

with open("./data/gg_region.json") as fr:
    geo_data = json.loads(fr.read())
```

```python
folium.Choropleth(
    geo_data=geo_data,
    data=population_df[target_year],
    columns=[population_df.index, population_df[target_year]],
    threshold_scale=[10000, 100000, 300000, 500000, 700000],
    key_on="feature.properties.name",
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.3
).add_to(gg_map)
```

```python
gg_map
```
