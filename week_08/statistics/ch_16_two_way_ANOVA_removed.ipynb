# ch 16. two-way ANOVA 

이전 챕터에서 독립 변수가 범주형 변수 1개, 종속 변수가 연속형 변수 1개 일 때 사용하 가능한 one-way ANOVA에 대해서 배워보았습니다. 이번 챕터에서는 이를 확장한 two way ANOVA에 대해서 알아 보겠습니다.

이름에서 짐작할 수 있듯, two-way ANOVA에서는 종속 변수에 영향을 주는 독립 변수가 2개입니다. 아래 예시는 자동차 충돌 실험에서 자동차 크기에 따른 4가지 등급과 자동차 제조사별 마네킹 머리의 파손 정도를 조사한 것입니다.

|  | 소형 자동차 | 준중형 자동차 | 중형 자동차 | 대형 자동차 |
| --- | --- | --- | --- | --- |
| A사 평균 | 760 | 610 | 520 | 530 |
| B사 평균 | 770 | 620 | 530 | 540 |

위 표만 보면 인형의 파손 정도는 자동차 등급에 따라 차이가 나는 듯 합니다. 제조사에 의해서도 차이가 나는 듯 하지만, 유의한 수준까지는 아닌 것으로 보입니다.

각 독립 변수가 종속 변수에 미치는 영향을 main effect라고 부릅니다. 그런데 독립 변수끼리도 서로 영향을 준다면 어떨까요? 가령 다른 독립 변수의 수준에 따라서, 다른 독립변수가 종속 변수에 미치는 영향이 달라진다면 어떨까요? 이를 interaction이라고 부릅니다. two-way ANOVA에서는 이 interaction을 고려해주어야 합니다. 

## interaction

서로 다른 두 독립변수가 어떤 상황에서 interaction이 있는지 예시를 통해서 알아보겠습니다.

(a) 경우에는 요인 1의 수준에 따라서 평균값이 증가하고 있습니다만, 요인 2에 의해서는 유의미한 차이를 보여주지 못합니다. 즉, 요인 1의 main effect가 존재하고, 둘 사이의 interaction은 없습니다.

(b) 경우에는 요인 2만 평균값에 유의미한 영향을 주고 있습니다. 

(c)의 경우, 요인 1과 요인2가 모두 종속 변수에 영향을 미치고 있습니다. 하지만 이 때에도 한 독립 변수의 그룹이 다른 독립 변수가 평균값에 미치는 영향에 영향을 주지 않습니다. 

(d) 경우에는 요인 1과 요인 2가 모두 종속 변수에 영향을 줍니다. 한 요인의 main effect가 다른 요인의 그룹에 의해서 달라지는 경우를 interaction이 존재한다고 합니다.

## two-way ANOVA

main effect와 interaction에 대해서 알아보았으니, 이제 two-way ANOVA에서의 가설 설정, 통계량, 분포, p-value 계산과 통계적 의사결정 과정에 대해서 알아보겠습니다.

###  가설 설정

two-way ANOVA에서는 세 가지 가설을 세웁니다.

**가설 1**  
귀무 가설: 요인 1과 요인 2는 interaction이 없다.   
대립 가설: 요인 1과 요인 2는 interaction이 있다.  

**가설 2**  
귀무 가설: 요인 1의 그룹별 평균은 같다.  
대립 가설: 요인 1의 그룹별 평균은 다르다.  

**가설 3**  
귀무 가설: 요인 2의 그룹별 평균은 같다.  
대립 가설: 요인 2의 그룹별 평균은 다르다.  

먼저 두 요인 사이에 interaction이 있는지 없는지에 대한 가설 검정을 진행합니다. 만약 interaction이 있다면 거기서 멈추고, 없다면 각 요인별 가설 검정과 사후 분석을 진행하면 됩니다. 

### 검정 통계량 계산

미리 말씀드리지만, two-way ANOVA의 계산 과정은 복잡합니다. 수식을 모두 암기해야한다는 생각은 내려놓고, 이런 방식으로 계산하는구나 정도로 이해하고 넘어가면 됩니다. 핵심만 정리하면 각 독립 요인 별로 f 통계량을 계산하고, 둘 사이의 interaction에 대한 f 통계량을 추가로 계산한다고 생각하면 됩니다. 

각 f 통계량 계산 수식과 풀이 과정이 궁금하신 분들은 아래 링크를 참고해주세요.  
https://nate9389.tistory.com/1635

$$f-value_{\text{interaction effect}}=\frac{Between\;Variance_{interaction\;effect}}{Within\;Variance}$$

$$f-value_{\text{main effect 1}}=\frac{Between\;Variance_{main\;effect 1}}{Within\;Variance}$$

$$f-value_{\text{main effect 2}}=\frac{Between\;Variance_{main\;effect 2}}{Within\;Variance}$$

### p-value 계산 및 통계적 의사결정

계산한 f 값들은 각각의 자유도와 f 분포표를 이용해서 p-value를 계산합니다. 이 때 순서가 중요합니다. 먼저 두 요인 사이에 interaction이 있는지 여부를 먼저 검정합니다. 만약 유의 하다면, 여기서 멈추고 상호 효과에 대한 사후 분석을 진행합니다.

상호 효과가 없다는 가설을 기각하지 못했다면, 각 요인에 대한 가설들을 검정하면 되며, one-way ANOVA와 마찬가지로 사후 분석을 진행하면 됩니다.

## two-way ANOVA 실습

이전 챕터에서 사용했던 데이터 셋을 가지고 PaymentMethod와 Contract 두 가지 독립변수가 TotalCharges에 미치는 영향에 대한 two-way ANOVA를 진행해보겠습니다.

### interaction 시각화

먼저 두 독립변수 사이에 상호 작용이 있는지 그래프를 그려서 확인해보겠습니다. seaborn의 pointplot을 이용하면 편리합니다.

PaymentMethod가 Electronic Check인지, Mailed Check 인지에 따라서 TotalCharges가 달라집니다. 그런데 달라지는 정도가 Month-to-month인지, One year인지, Two year인지에 따라서 영향을 받습니다. 즉, 두 독립변수 사이에 상호 작용이 있어 보이지만, 통계적으로 유의한 수준인지는 아직 알 수 없습니다. two-way ANOVA를 사용해서 검정을 해보겠습니다.

scipy와 사용법이 많이 다릅니다만 이해하기 크게 어렵지 않습니다. 먼저 C(PaymentMethod)와 C(Contract)는 각각 독립 변수 PaymentMethod와 Contract를 의미합니다. 그리고 C(PaymentMethod):C(Contract)는 이 둘의 상호 작용을 의미합니다. 종속 변수 TotalCharges는 2개의 독립 변수와 이 둘의 상호 작용으로 설명됩니다.

책 마다 소개되는 기법이 다릅니다만, 이 경우 각 독립 변수별 사후 분석은 진행하지 않고 분석을 멈춥니다. 그리고 시각화 자료를 이용해서 두 독립 변수의 상관 관계를 설명하고, 원인을 파악하는 선에서 분석을 멈춥니다.

만일 두 독립 변수 사이 상호 작용 효과가 없다면, one-way ANOVA와 동일하게 Tukey 기법을 사용해서 사후 분석을 진행할 수 있습니다. 혹은 각 독립 변수 별로 one-way ANOVA를 진행할 수도 있습니다.

## 정리
지금까지 one-way ANOVA, two-way ANOVA에 대해서 살펴보았습니다. 이외에도 ANOVA에는 다양한 종류가 있습니다. 주어진 데이터 셋에 적합한지 면밀히 검토하고 사용하면 좋겠습니다.

- 삼원 배치 분산 분석(Three-way ANOVA):
    - 독립 변수가 3개 이상일 때 사용. 해석이 어렵기 때문에 잘 사용 안함.
- 반복 측정 분산 분석(Repeated Measure ANOVA):
    - 같은 집단에 대해 여러번 측정을 한 데이터의 평균이 같은지 검정. paired t-test는 전과 후 2번 측정했을 때만 사용이 가능. 3번 이상은 repeated measure ANOVA를 사용
