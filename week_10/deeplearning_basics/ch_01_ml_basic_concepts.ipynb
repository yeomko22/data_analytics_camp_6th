# ch 1. ML 기본 개념 정리

## Machin Learning
- explicit programming: 개발자가 일일이 룰을 코딩하는 방식의 프로그래밍
- machine learning: 데이터로부터 알고리즘이 스스로 룰을 찾는 방식의 프로그래밍

예시) 스팸 메일 분류기  
- explicit programming: 특정 단어가 포함되면 스팸으로 분류하는 등의 룰들의 집합으로 표현  
- machine learning: 스팸인지 아닌지 분류되어 있는 데이터를 학습해서 스팸 메일을 분류하는 룰을 알아서 찾아냄.

## Supervised Learning vs Unsupervised Learning

- Supervised Learning
    - 라벨이 붙어있는 데이터 셋을 이용해 알고리즘을 학습하는 방법
    - 데이터 셋 구축에 많은 인력과 비용이 필요함
    - ex) 타이타닉 생존자 예측
- Unsupervised Learning
    - 라벨이 안붙어있는 데이터 셋을 이용해 알고리즘을 학습하는 방법
    - 데이터 셋 구축이 쉬움.
    - ex) 뉴스 클러스터링
    
딥러닝 등장한 이후로 오랜 기간동안 라벨이 붙어있는 데이터 셋을 이용해 모델을 학습시키는 Supervised Learning이 대세였으며, 여전히 많이 활용됩니다. Supervised Learning을 위해선 라벨이 붙어있는 데이터 셋이 필요하며, 비용과 시간이 많이 들어갑니다.

Unsupervised Learning은 비교적 많이 활용되지 못했던 기법이었습니다. clustering 정도 분야에서 활용되었습니다만 그 외엔 마땅한 적용 분야가 없었습니다. 그런데 자연어 처리 분야에서 openAI가 Unsupervised Learning 방식으로 학습시킨 GPT 모델들을 발표하면서 급부상 중입니다.

## Classification vs Regression

- Classification
    - 입력 데이터를 바탕으로 종류를 구분하는 모델 (종속 변수가 범주형 변수)
    - 분류를 잘해내는 것이 중요 
- Regression
    - 입력 데이터를 바탕으로 특정한 값을 예측하는 모델 (종속 변수가 연속형 변수)
    - 특정 값을 잘 맞추는 것이 중요

## Overfitting vs Underfitting

- Underfit
    - 모델이 너무 단순해서 데이터 셋을 충분히 설명하지 못하는 것
- Overfit
    - 학습 데이터에 지나치게 과적합 되어 일반적인 설명력이 떨어지는 것



## Bias vs Variance

Underfitting, Overfitting과 관련해서 알아두어야 할 개념이 있습니다. 바로 Bias, Variance tradeoff 입니다.

- Bias: 예측값과 예측하려는 실제 데이터 값 사이의 차이입니다.
- Variance: 모델의 예측값의 변동성입니다.



우리가 원하는 것은 실제 데이터를 정확히 예측하면서도, 예측값의 변동성이 작은 모델입니다. 즉, bias와 variance가 모두 작은 모델을 원합니다. 하지만 bias와 Variance은 한 쪽에 집중하면 다른 한 쪽이 떨어지는 tradeoff 관계에 있습니다.

먼저 Underfitting된 모델을 가정하겠습니다. 이는 예측해야하는 값에서 많이 벗어난 예측을 내리게 됩니다. 즉, 위 그래프에서 High Bias, Low Variance에 해당합니다. 이런 경우 파라미터를 더 추가하여 모델이 데이틀 충분히 설명할 수 있도록 해야합니다.

반대로 Overfitting된 모델을 가정하겠습니다. 이는 학습 데이터 셋에 지나치게 과적합되어, 실제 예측값을 내릴 때에 변동성이 커지게 됩니다. Overfitting된 분류 모델이 구불구불한 선 모양임을 떠올려보면 분산이 크다는 것을 직관적으로 이해하기 쉽습니다.

즉, 모델이 Underfitting 될 경우 Bias가 커지고, Overfitting 될 경우 Variance가 커지게 됩니다. 우리가 원하는 것은 이런 tradeoff 관계에서 bias와 variance가 모두 작은 모델을 찾는 것입니다.



## Tensorflow vs Pytorch

딥러닝을 구현하기 위해서는 복잡한 신경망과 이를 학습시킬 수 있는 오류 역전파 알고리즘, 데이터 파이프라인 구축, GPU 가속 등의 굉장히 복잡한 기능들이 필요합니다. 이를 직접 구현하지는 않고, pytorch라는 딥러닝 프레임워크를 사용할 예정입니다.

현재 딥러닝 모델 학습 프레임워크는 구글의 tensorflow와 pytorch가 양분하고 있습니다만, 점점 더 pytorch 쪽으로 무게가 실리고 있습니다. 아래 그래프는 Paper with Code라는 사이트에서 발표한 AI 논문 구현에 사용된 프레임워크 비율입니다. 




딥러닝 열풍이 불던 초기에는 tensorflow가 압도적으로 많이 사용되었습니다만, pytorch가 점점 인기를 끌더니 현재는 대부분의 논문 구현체들이 pytorch를 사용합니다. 

그 이유로는 여러가지가 있습니다만, pytorch의 가장 큰 장점은 가장 python 스럽게 딥러닝 모델을 개발할 수 있다는 점입니다. tensorflow를 포함하여 다른 프레임워크들은 언어는 python을 사용하지만 API들이 마치 다른 언어를 사용하는 것 같은 불편함이 있었습니다. pytorch는 최대한 python 기존 문법들을 사용하여 개발이 가능하여 많은 개발자들의 인기를 끌었습니다.

최근 데이터, AI 관련 직군의 채용 공고를 보면 요구되는 기술 스택으로 pytorch가 tensorflow에 비해 압도적으로 많이 보입니다. 때문에 우리 수업에서는 pytorch를 선택하여 실습을 진행할 예정입니다. (물론 기본 개념을 확실히 익히면 tensorflow도 금방 배웁니다.)
