# ch 2. R square 

지난 챕터에서는 단순 선형 회귀를 이용해서 집 값을 예측하는 모델을 학습시켜 보았습니다. 그리고 그 결과를 표로 출력해보았습니다. 이번 챕터에서는 선형 모델의 적합도를 평가하는 지표 중 하나인 R square에 대해서 알아보겠습니다.

## 모델 준비

## Sum of Squares

r-square를 이해하기 위해서는 먼저 SST, SSR, SSE에 대해서 알아야합니다. 그림으로 먼저 이해해보겠습니다.



먼저 파란색 점들은 각각의 데이터 샘플들입니다. 빨간색 점선으로 표기된 것은 종속 변수의 평균 값으로 보통 Ȳ로 표현되는 값입니다. 보라색 점선은 최소제곱법을 통해서 구한 회귀식입니다.

### SST
먼저 SST는 total sum of squares의 약자로 각각의 종속 변수에서 종속 변수들의 평균 값을 빼준 값들의 제곱합을 의미합니다.

$$SST=\sum_{i=1}^{n}{(y_i-\bar{y})}^2$$

### SSR
그 다음, SSR은 sum of squares due to regression의 약자로, 회귀식이 예측한 값과 종속 변수의 평균 값 사이의 차이의 제곱합을 의미합니다. 이는 전체 변동에서 회귀식에 의해 설명되는 부분이라고 볼 수 있습니다. 

$$SSR=\sum_{i=1}^{n}{({\hat{y}}_i-\bar{y})}^2$$

### SSE
마지막으로 SSE는 error sum of squares의 약자로, 실제 관측값과 회귀식이 예측한 값 사이의 차이의 제곱합입니다. 이는 전체 변동에서 회귀식이 설명해내지 못하는 부분이라고 볼 수 있습니다.

$$SSE=\sum_{i=1}^{n}(y_i-{\hat{y}}_i)^2$$

SST, SSR, SSE 세 값의 관계를 수식으로 표현하면 아래와 같습니다.

$$SST=SSR+SSE$$

$$\sum_{i=1}^{n}(y_i - \bar{y})^2=\sum_{i=1}^{n}{({\hat{y}}_i-\bar{y})}^2+\sum_{i=1}^{n}(y_i-{\hat{y}}_i)^2$$

이를 해석해보면, 전체 데이터 샘플의 평균으로부터 떨어져 있는 변동량(SST)은 회귀식이 설명해내는 부분(SSR)과 회귀식이 설명하지 못하는 부분(SSE)으로 나눌 수 있다는 것입니다.

## R square
### R square 정의

흩어져 있는 자료들을 하나의 직선으로 정리한 것이 회귀식입니다. 그렇다면 회귀식이 원래의 자료를 잘 반영하고 있는지 하나의 계수로 나타낼 수 있으면 편리하겠죠?  R square는 데이터의 총 변동량 중 회귀식이 설명하는 비율이 얼만큼인지를 나타내주는 지표입니다. 

r-square의 수식은 아래와 같습니다.

$$R^2=\frac{SSR}{SST}=1-\frac{SSE}{SST}=\frac{\sum_{i=1}^{n}(y_i-{\hat{y}}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}$$

이렇게 계산한 r-square 값은 0과 1 사이 값을 가집니다.

- 1에 가깝다 → 추정된 회귀식이 총 변동량의 많은 부분을 설명한다.
- 0에 가깝다 → 추정된 회귀식이 총 변동량을 적절하게 설명하지 못한다.

집 값 예측 회귀 모델에서 R square 값은 0.493이었습니다. 이는 전체 데이터 변동의 49% 정도를 회귀식이 설명하는 것입니다.

### 적정 R square 값

한편 R square 값이 어느정도가 적절하다라는 판단 기준은 도메인마다 다릅니다. 사회 현상의 경우 80%가 넘어도 상당히 큰 값으로 간주되고, 경영전략 분야에서는 30% 정도인 것도 있습니다.

단순히 R 제곱 값만으로 모델을 평가하기보다는 다른 지표들과 함께 고려하는 것이 중요합니다. 예를 들어, 잔차 분석, 오차의 정규성 검정, 다중공선성 등을 평가하는 것이 모델의 성능을 정확히 판단하는 데 도움이 됩니다. 또한, 도메인 지식과 문제의 특성을 고려하여 모델을 평가하는 것이 중요합니다.

## 정리
